```python
import os
import re
import numpy as np
import pandas as pd
import csv
from xml.etree import ElementTree as ET

from scripts.word_embeddings import load_embedding_weights
```

```python
path = '/home/petar/Fakultet/Semester 7/NLP/Datasets/fake_news/train/en'
labels = '/home/petar/Fakultet/Semester 7/NLP/Datasets/fake_news/train/truth_en.txt'
```

```python
with open('file1.csv', 'w') as f:
    writer = csv.writer(f)
    print('1')
    for file in os.listdir(path):
        
        temp_file = path + '/' + file
        
        tree = ET.parse(temp_file)
        
        print('2')
    
        for document in tree.find('documents').findall('document'):
            
            
            print('3')
            
            text = document.text.replace('amp;', '')
            
            text = re.sub('[;]', '', text)

            writer.writerow([file.replace('.xml', ''), text])
```

```python
f.close()
```
